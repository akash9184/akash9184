```python
import pandas as pd
import sqlite3
import datetime
```

### Q1

```python
connection=sqlite3.connect('Company.db')
con=connection.cursor()
```

```python
get_records="Select name from sqlite_master where type='table';"
con.execute(get_records)
print(con.fetchall())
```

```python
df=pd.read_sql_query("SELECT * FROM Employee",connection)
df
```

```python
df=pd.read_sql_query("SELECT * FROM Bonus",connection)
df
```

```python
df=pd.read_sql_query("SELECT Employee.EMPLOYEE_ID,EMPLOYEE.FIRST_NAME || ' ' || EMPLOYEE.LAST_NAME AS Name,SUM(Bonus.BONUS_AMOUNT) AS Total_Bonus FROM Employee INNER JOIN Bonus ON Employee.EMPLOYEE_ID=Bonus.EMPLOYEE_REF_ID GROUP BY EMPLOYEE_ID",connection)
df
```

```python
df=pd.read_sql_query("SELECT salesperson.name as Salesperson,Count(orders.salesperson_id) as 'Number of Orders' from salesperson inner join orders on salesperson.id = orders.salesperson_id group by salesperson_id having Count(orders.salesperson_id)>=2 order by Count(orders.salesperson_id) DESC ",connection)
df
```

```python
df=pd.read_sql_query("SELECT Count(attendance_events.student_id) * 100 / (select count(student_id) from attendance_events) as Percent from attendance_events join all_students on all_students.student_id = attendance_events.student_id where strftime('%m',attendance_events.date_event)=strftime('%m',all_students.date_of_birth) and strftime('%d',attendance_events.date_event)=strftime('%d',all_students.date_of_birth) and attendance_events.attendance='present' ",connection)
df
```
